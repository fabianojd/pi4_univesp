```python
# Carregar bibliotecas

import pandas as pd
import numpy as np
import zipfile
import requests
from io import BytesIO
import os
```

```python
# criar uma pasta para gravar o datset
os.makedirs("./dataset", exist_ok=True)
```

```python
# Copia base de dados para Arquivo do Google Colab
# https://dadosabertos.camara.leg.br/swagger/api.html#staticfile

url = 'https://www.camara.leg.br/cotas/Ano-2019.csv.zip'

filebytes = BytesIO(
    requests.get(url).content
)

meuzip = zipfile.ZipFile(filebytes)
meuzip.extractall("./dataset")
```

```python
# Copia base de dados para Arquivo do Google Colab
# https://dadosabertos.camara.leg.br/swagger/api.html#staticfile

url = 'https://www.camara.leg.br/cotas/Ano-2022.csv.zip'

filebytes = BytesIO(
    requests.get(url).content
)

meuzip = zipfile.ZipFile(filebytes)
meuzip.extractall("./dataset")
```

```python
dados_2019 = pd.read_csv(
    "./dataset/Ano-2019.csv", sep = ";"
)
dados_2020 = pd.read_csv(
    "./dataset/Ano-2020.csv", sep = ";"
)
dados_2021 = pd.read_csv(
    "./dataset/Ano-2021.csv", sep = ";"
)
dados_2022 = pd.read_csv(
    "./dataset/Ano-2022.csv", sep = ";"
)
```

```python
dados = pd.concat([dados_2019, dados_2020, dados_2021, dados_2022])
```

```python
# Descarte de colunas desnecessárias
dados.drop(["cpf", "nuCarteiraParlamentar", "nuLegislatura", "codLegislatura", "numEspecificacaoSubCota", "txtDescricaoEspecificacao", "txtFornecedor", "txtCNPJCPF", "txtNumero", "indTipoDocumento", "datEmissao", "vlrGlosa", "vlrLiquido", "numParcela", "txtPassageiro", "txtTrecho", "numLote", "numRessarcimento", "vlrRestituicao", "nuDeputadoId", "ideDocumento", "urlDocumento"], axis=1, inplace=True)
```

```python
dados
```

```python
# Renomear colunas
dados.rename({"txNomeParlamentar": "Nome Parlamentar"}, axis=1, inplace=True)
dados.rename({"ideCadastro": "id Cadastro"}, axis=1, inplace=True)
dados.rename({"sgUF": "UF"}, axis=1, inplace=True)
dados.rename({"sgPartido": "Partido"}, axis=1, inplace=True)
dados.rename({"numSubCota": "Código gastos"}, axis=1, inplace=True)
dados.rename({"txtDescricao": "Tipo de gastos"}, axis=1, inplace=True)
dados.rename({"vlrDocumento": "Valor gasto"}, axis=1, inplace=True)
dados.rename({"numMes": "Mês"}, axis=1, inplace=True)
dados.rename({"numAno": "Ano"}, axis=1, inplace=True)
```

```python
# Exportar CSV
dados.to_csv('./dataset/Base_de_dados_2019a2022.csv', index=False, sep=";")  
```
